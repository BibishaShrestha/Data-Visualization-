## Data Visualization
### Data table
Health_heart_experiment

```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/Health_heart_experimental.csv")
df.head(5)
```

### Line Chart
Age vs Systolic Blood Pressure

```python
data = {
    "age": [64, 21, 30, 35, 39, 68, 50, 23, 34, 36],
    "SysBP": [141, 109, 112, 106, 140, 115, 125, 111, 144, 113]
}

df = pd.DataFrame(data)

age = df["age"].tolist()
sysbp = df["SysBP"].tolist()

plt.plot(sysbp, age, color='blue', marker='o', linestyle='dashed')

plt.title("Age vs Systolic Blood Pressure")
plt.xlabel("Systolic Blood Pressure (SysBP)")
plt.ylabel("Age (Years)")

plt.show()
```

Age vs Diastolic Blood Pressure

```python
data = {
    "age": [56, 21, 40, 35, 39, 68, 50, 63, 34, 36],
    "DiaBP": [96, 100, 73, 90, 90, 96, 117, 85, 80, 109]
}

df = pd.DataFrame(data)

sns.lineplot(x='age', y='DiaBP', data=df)

plt.title("Age vs Diastolic Blood Pressure")
plt.xlabel("Age (Years)")
plt.ylabel("Diastolic Blood Pressure (DiaBP)")

plt.show()
```

### Bar Plot
Diastolic Blood Pressure by Age

```python
age = [21, 23, 30, 34, 35, 36, 39, 50, 64, 68]
diabp = [100, 85, 73, 80, 90, 109, 90, 117, 96, 96]

plt.bar(range(len(age)), diabp, color='skyblue')

plt.title("Diastolic Blood Pressure by Age")
plt.ylabel("Diastolic Blood Pressure (DiaBP)")
plt.xlabel("Individuals (Sorted by Age)")

plt.xticks(range(len(age)), age)

plt.show()
```

Avg. age by sex and weight

```python
data = {
    'age': [44, 21, 35, 37, 39, 58, 40, 23, 34],
    'sex': [1, 1, 0, 1, 0, 1, 1, 1, 1],
    'weightKg': [69, 48, 65, 45, 92, 53, 56, 55, 67],
}
df = pd.DataFrame(data)

sns.barplot(x='weightKg', y='age', hue='sex', data=df)
plt.title('Average Age by Sex and weight')
plt.xlabel('Weight (kg)')
plt.ylabel('Average Age')
plt.show()
```

### Piechart

```python
data = {

    'sex': [1, 1, 0, 1, 0, 1, 1, 1, 1]
}
df = pd.DataFrame(data)

df['sex'].value_counts().plot(kind='pie', autopct='%0.2f', labels=['Male', 'Female'],colors=['lightgreen', 'pink'])
plt.title('Chart of proportion of sex')
plt.show()
```

### Box plot

For Age:

```python
data = {'age': [44, 21, 35, 37, 39, 58, 40, 23, 34]}
df = pd.DataFrame(data)

sns.boxplot(df['age'])
plt.title('Boxplot of Age')
plt.xlabel('Age')
plt.ylabel("")
plt.show()
```

 Age distribution by sex

```python
data = {
    'age': [44, 21, 35, 37, 39, 58, 40, 23, 34],
    'sex': [1, 1, 0, 1, 0, 1, 1, 1, 1],
}
df = pd.DataFrame(data)

df['sex'] = df['sex'].map({1: 'Male', 0: 'Female'})

sns.boxplot(x='sex', y='age', data=df)
plt.title('Age Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

### Histogram


Distribution Of Age

```python
data = {'age': [44, 21, 35, 37, 39, 58, 40, 23, 34]}
df = pd.DataFrame(data)

sns.displot(df['age'], kind="hist", color='lightgreen')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Distribution of Age with KDE

```python
data = {'age': [44, 21, 35, 37, 39, 58, 40, 23, 34]}
df = pd.DataFrame(data)

sns.displot(df['age'], kind="hist", kde=True, color='lightgreen')
plt.title('Distribution of Age with KDE')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## Bivarient analysis

```python
data = {
    'age': [65, 31, 20, 35, 49, 68, 50, 23, 34],
    'indication': [1, 0, 0, 0, 1, 0, 0, 0, 0]
}
df = pd.DataFrame(data)

sns.kdeplot(df[df['indication'] == 0]['age'], label='No Indication', color='red')
sns.kdeplot(df[df['indication'] == 1]['age'], label='Indication', color='green')

plt.title('Age Distribution by Indication')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```
